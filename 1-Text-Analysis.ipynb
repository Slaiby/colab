```python
#Count how often a word occurs in a text, and compute what percentage of the text is taken up by a specific word:

count = text3.count("God") # this is case sensitive!

perc = 100 * text3.count('God') / len(text3)

print("count of word God is: ", count)
print("percentage of word God is: ", perc)
```

```python
def lexical_diversity(text):
    return len(text) / len(set(text))
```

```python
def percentage(count, total): 
    return 100 * count / total
```

```python
print(lexical_diversity(text3))

print(percentage(count, len(text3)))
```
